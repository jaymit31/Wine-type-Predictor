```python
import pandas as pd
```

```python
white=pd.read_csv("winequality-white-kaggle.csv",sep=";")
```

```python
white["type"]="white"
```

```python
red=pd.read_csv("winequality-red-kaggle.csv",sep=";")
```

```python
red["type"]="red"
```

```python
white.head()
```

```python
red.head()
```

```python
sample=red.append(white,ignore_index=True)
```

```python
sample.loc[1,:]
```

```python
 sample
```

```python

sample1 = sample.drop(['citric acid', 'residual sugar', 'density', 'pH', 'quality'], axis = 1)
sample1
```

# 1. Dropping the Output Column

```python
X=sample1.drop("type",axis=1)
```

```python
y=sample1["type"]
```

# 2. Splitting the Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=.1, stratify=y)
```

# 3. Training

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=RandomForestClassifier(n_estimators = 100)
```

```python
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

# 4. Accuracy

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
